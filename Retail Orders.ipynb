```python
#Extract file from zip file
import zipfile
zip_path=r"D:\Data Analysis\SQL and Python Project\orders.csv.zip"
zip_ref=zipfile.ZipFile(zip_path)
zip_ref.extractall() #Extract files to dir
zip_ref.close() #close file
```

```python
#read data from the file and handle null values
import pandas as pd
file_path=r"D:\Data Analysis\SQL and Python Project\orders.csv"
df=pd.read_csv(file_path)
df.head(20)
```

```python
#to get distinct value available in "shipmode"column since we have "Not available" and "unknown" values present in the column
df['Ship Mode'].unique()
```

```python
#since there are 'Not Available', 'unknown', that values also should be treated as null.To make that null below code is used
df=pd.read_csv(file_path,na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#rename column names .. make them lower case and replace space with underscore
#df.rename(columns={'Order Id':'order_id','City:''city'}) #this is one way to change using JSON format. but it is not applicable if there are so many columns
#Effective Method
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(2)
```

```python
#derive new columns discount,sale price and profit
df['discount']=df['list_price']*df['discount_percent']*.01  #discount calculation
df['sale_price']=df['list_price']-df['discount']  #sale price calculation
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
#convert order date from object data type to datetime
df.dtypes #display the datatype pf each column
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d') #convert to datetime format
df
```

```python
#Drop cost price, list price and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
